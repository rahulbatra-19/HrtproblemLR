```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
hrt = pd.read_csv('framingham.csv')
hrt.head(n=10)
```

```python
column = ['male','age','education','currentSmoker','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totchol','sysBP','diaBP','BMI','heartRate','glucose','TenYearCHD']
hrt = pd.read_csv('framingham.csv',names=column ,header=1)
hrt.head()
```

```python
hrt.tail()
```

```python
hrt.count()
```

```python
hrt.describe()
```

```python
features = ['male','age', 'prevalentStroke','prevalentHyp','sysBP','diabetes','currentSmoker']
A = hrt[features]
B = hrt.TenYearCHD
```

```python
from sklearn.model_selection import train_test_split
A_train,A_test, B_train,B_test = train_test_split(A,B,random_state=0)
```

```python
logrg = LogisticRegression()
logrg.fit(A_train,B_train)
```

```python
B_pred = logrg.predict(A_test)
from sklearn.metrics import accuracy_score, recall_score
from sklearn.metrics import confusion_matrix
```

```python
accuracy_score(B_test, B_pred)
```

```python
print("True Ten year CHD:  "), B_test.values[70:80]
```

```python
print("Predicted Ten year CHD: "),B_pred[70:80]
```

```python
confusion_matrix(B_test,B_pred)
```
